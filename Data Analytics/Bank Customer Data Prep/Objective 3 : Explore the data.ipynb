```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bank_churn_df = pd.read_csv("Bank Customer Churn Data/Prepared Data/cleaned_bank_churn_df.csv",
                            encoding="ISO-8859-1",
                            low_memory=False)
bank_churn_df.info()
```

```python
bank_churn_df["Exited"].value_counts(normalize=True).plot.bar()
```

```python
sns.barplot(data=bank_churn_df,x = "Geography",y = "Exited")
```

```python
for col in bank_churn_df.drop("Surname", axis=1).select_dtypes("object"):
    plt.figure(figsize=(8, 4))
    sns.barplot(data=bank_churn_df, x=col, y="Exited")
    plt.title(f'Relationship between {col} and Exited')
    plt.xlabel(col)
    plt.ylabel("Exited")
    plt.show()
```

```python
for col in bank_churn_df.drop(["CustomerId","Exited"], axis=1).select_dtypes("number"):
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=bank_churn_df, y=col, hue="Exited")
    plt.title(f'Relationship between {col} and Exited')
    plt.xlabel(col)
    plt.ylabel("Exited")
    plt.show()
```

```python
for col in bank_churn_df.drop(["CustomerId","Exited"], axis=1).select_dtypes("number"):
    plt.figure(figsize=(8, 4))
    sns.histplot(data=bank_churn_df, x=col, hue="Exited",kde=True)
    plt.title(f'Relationship between {col} and Exited')
    plt.xlabel(col)
    plt.ylabel("Exited")
    plt.show()
```

```python
for col in bank_churn_df.drop("Surname", axis=1).select_dtypes("object"):
    plt.figure(figsize=(8, 4))
    sns.barplot(data=bank_churn_df, x=col, y="Exited")
    plt.title(f'Relationship between {col} and Exited')
    plt.xlabel(col)
    plt.ylabel("Exited")
    plt.show()
```
