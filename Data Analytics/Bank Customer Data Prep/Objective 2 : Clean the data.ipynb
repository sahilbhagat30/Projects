```python
import pandas as pd 
import numpy as np
```

```python
bank_churn_df = pd.read_csv("Bank Customer Churn Data/Prepared Data/bank_churn_df.csv",
                            encoding="ISO-8859-1",
                            low_memory=False)
bank_churn_df.info()
```

```python
bank_churn_df = bank_churn_df.drop("Unnamed: 0", axis = 1)
```

```python
bank_churn_df.head()
```

```python
bank_churn_df[bank_churn_df.isna().any(axis = 1)== True]
```

```python
bank_churn_df = bank_churn_df.fillna(value = {"Surname" : "MISSING", "Age" : bank_churn_df["Age"].median()})
bank_churn_df.info()
```

```python
bank_churn_df.iloc[[28,121,9389]]
```

```python
bank_churn_df["EstimatedSalary"] = bank_churn_df["EstimatedSalary"].replace(-999999.0, bank_churn_df["EstimatedSalary"].median())

bank_churn_df.describe()
```

```python
bank_churn_df["Geography"].value_counts()
```

```python
bank_churn_df["Geography"] = np.where(bank_churn_df["Geography"].isin(["FRA","French","France"]),"France",bank_churn_df["Geography"])
bank_churn_df["Geography"].value_counts()
```

```python
bank_churn_df.to_csv("Bank Customer Churn Data/Prepared Data/cleaned_bank_churn_df.csv")
bank_churn_df.describe()
```
