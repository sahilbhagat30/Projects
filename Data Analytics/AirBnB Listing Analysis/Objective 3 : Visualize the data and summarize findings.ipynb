```python
import pandas as pd
import seaborn as sns
```

```python
paris_listing_neighbourhood = pd.read_csv("Airbnb Data/Prepared Data/obj2_paris_listings_neighbourhood.csv",
                               encoding = "ISO-8859-1",
                               low_memory=False)
paris_listings_over_time = pd.read_csv("Airbnb Data/Prepared Data/obj2_paris_listings_over_time.csv",
                               encoding = "ISO-8859-1",
                               parse_dates=["host_since"],
                               low_memory=False)
paris_listing_neighbourhood.info()
paris_listings_over_time.info()
```

```python
(paris_listing_neighbourhood
 .plot
 .barh(
     title = "Average Listing Price by Price Neighbourhood",
     xlabel = "Preice Per Night (Euro)",
     ylabel = "Neighbourhood",
     legend = None
 ))
sns.despine
```

```python
(paris_listings_over_time.plot(
     x="host_since",  
     y="neighbourhood",  
     title = "New AirBnb Host in Paris Over Time",
     ylabel = "New Hosts",
     xlabel = "Year"
 ))
sns.despine()
```

```python
(paris_listings_over_time
 .plot(
     x = "host_since",
     y = "price",
     title = "Average AirBnb Price in Paris Over Time",
     ylabel = "Average Price",
     xlabel = "Years"
     
 ))
sns.despine
```
