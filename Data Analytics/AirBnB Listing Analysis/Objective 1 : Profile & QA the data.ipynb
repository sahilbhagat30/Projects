```python
import pandas as pd
```

```python
listings = pd.read_csv('Airbnb Data/Listings.csv', 
                       encoding='ISO-8859-1', 
                       low_memory=False,
                       parse_dates=["host_since"])
listings.head()
```

```python
listings.info()
```

```python
paris_listing = (
                listings
                .query("city == 'Paris'")
                .loc[:,["host_since","neighbourhood","city","accommodates", "price"]]
                )
paris_listing.info()
```

```python
paris_listing.isna().sum()
```

```python
paris_listing.describe()
```

```python
paris_listing.query("accommodates == 0 and price == 0").count()
```

```python
paris_listing.to_csv('Airbnb Data/Prepared Data/obj1_paris_listings.csv')
```
